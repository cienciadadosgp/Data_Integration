```python
# URL do arquivo XML
url = "https://raw.githubusercontent.com/cienciadadosgp/Data_Integration/main/XML_Soc.xml"

# Carregar XML em DataFrame
df = pd.read_xml(url)

# Visualizar as primeiras linhas
df.head()
```

```python
# 1️⃣ Total de famílias cadastradas por ano/mês
total_familias = df.groupby(['ano_cadunico', 'mes_cadunico'])['total_numero_familia_cadunico'].sum().reset_index()
print("Total de famílias por ano/mês:")
display(total_familias)
```

```python
# Garantir que todos os valores sejam string antes de substituir vírgula
df['renda'] = df['renda'].astype(str).str.replace(',', '.', regex=False)

# Converter para float, valores inválidos viram NaN
df['renda'] = pd.to_numeric(df['renda'], errors='coerce')

# Calcular média, mediana e desvio
renda_stats = df.groupby(['ano_renda', 'mes_renda'])['renda'].agg(['mean','median','std']).reset_index()
display(renda_stats)
```

Análises de vulnerabilidade

Quantas famílias em situação de vulnerabilidade alta/média/baixa

Relação renda vs número de benefícios

Identificação de famílias em risco (potenciais candidatos a programas sociais)

```python
# Garantir que todos os valores sejam string antes de substituir vírgula
df['rendapercapta'] = df['rendapercapta'].astype(str).str.replace(',', '.', regex=False)

# Converter definitivamente para número (float)
df['rendapercapta'] = pd.to_numeric(df['rendapercapta'], errors='coerce')

# Verificar se agora é numérica
print(df['rendapercapta'].dtype)

# Função de classificação
def classificar_vulnerabilidade(renda):
    if pd.isna(renda):
        return 'desconhecido'
    elif 0 <= renda <= 218:
        return 'baixa'
    elif 219 <= renda <= 706:
        return 'media'
    elif renda >= 707:
        return 'alta'
    else:
        return 'desconhecido'

# Aplicar classificação
df['vulnerabilidade_social'] = df['rendapercapta'].apply(classificar_vulnerabilidade)

# Contar total de famílias por categoria
totais = (
    df['vulnerabilidade_social']
    .value_counts()
    .reindex(['baixa', 'media', 'alta'], fill_value=0)
    .reset_index()
)

totais.columns = ['vulnerabilidade_social', 'total_familias']

# Exibir resultado
print(totais)
```

```python
# Criar coluna 'risco' baseado na vulnerabilidade
totais['risco'] = totais['vulnerabilidade_social'].apply(lambda x: 0 if x == 'baixa' else 1)

# Criar rótulo legível
totais['risco_label'] = totais['risco'].map({0: 'Alto risco', 1: 'Baixo risco'})

# Exibir tabela final com total de famílias por risco
totais_final = totais[['vulnerabilidade_social', 'total_familias', 'risco_label']]
print(totais_final)
```

```python
import matplotlib.pyplot as plt

# Somar total de famílias por risco
totais_risco = totais.groupby('risco_label')['total_familias'].sum().reset_index()

# Plotar gráfico de barras
plt.figure(figsize=(6,4))
plt.bar(totais_risco['risco_label'], totais_risco['total_familias'], color=['red', 'green'])
plt.title('Total de Famílias por Risco')
plt.xlabel('Risco')
plt.ylabel('Número de Famílias')
plt.xticks(rotation=0)
plt.show()
```
