```python
# URL do arquivo XML
url = "https://raw.githubusercontent.com/cienciadadosgp/Data_Integration/main/XML_Soc.xml"

# Carregar XML em DataFrame
df = pd.read_xml(url)

# Visualizar as primeiras linhas
df.head()
```

```python
# 1ï¸â£ Total de famÃ­lias cadastradas por ano/mÃªs
total_familias = df.groupby(['ano_cadunico', 'mes_cadunico'])['total_numero_familia_cadunico'].sum().reset_index()
print("Total de famÃ­lias por ano/mÃªs:")
display(total_familias)
```

```python
# Garantir que todos os valores sejam string antes de substituir vÃ­rgula
df['renda'] = df['renda'].astype(str).str.replace(',', '.', regex=False)

# Converter para float, valores invÃ¡lidos viram NaN
df['renda'] = pd.to_numeric(df['renda'], errors='coerce')

# Calcular mÃ©dia, mediana e desvio
renda_stats = df.groupby(['ano_renda', 'mes_renda'])['renda'].agg(['mean','median','std']).reset_index()
display(renda_stats)
```

AnÃ¡lises de vulnerabilidade

Quantas famÃ­lias em situaÃ§Ã£o de vulnerabilidade alta/mÃ©dia/baixa

RelaÃ§Ã£o renda vs nÃºmero de benefÃ­cios

IdentificaÃ§Ã£o de famÃ­lias em risco (potenciais candidatos a programas sociais)

```python
# Garantir que todos os valores sejam string antes de substituir vÃ­rgula
df['rendapercapta'] = df['rendapercapta'].astype(str).str.replace(',', '.', regex=False)

# Converter definitivamente para nÃºmero (float)
df['rendapercapta'] = pd.to_numeric(df['rendapercapta'], errors='coerce')

# Verificar se agora Ã© numÃ©rica
print(df['rendapercapta'].dtype)

# FunÃ§Ã£o de classificaÃ§Ã£o
def classificar_vulnerabilidade(renda):
    if pd.isna(renda):
        return 'desconhecido'
    elif 0 <= renda <= 218:
        return 'baixa'
    elif 219 <= renda <= 706:
        return 'media'
    elif renda >= 707:
        return 'alta'
    else:
        return 'desconhecido'

# Aplicar classificaÃ§Ã£o
df['vulnerabilidade_social'] = df['rendapercapta'].apply(classificar_vulnerabilidade)

# Contar total de famÃ­lias por categoria
totais = (
    df['vulnerabilidade_social']
    .value_counts()
    .reindex(['baixa', 'media', 'alta'], fill_value=0)
    .reset_index()
)

totais.columns = ['vulnerabilidade_social', 'total_familias']

# Exibir resultado
print(totais)
```

```python
# Criar coluna 'risco' baseado na vulnerabilidade
totais['risco'] = totais['vulnerabilidade_social'].apply(lambda x: 0 if x == 'baixa' else 1)

# Criar rÃ³tulo legÃ­vel
totais['risco_label'] = totais['risco'].map({0: 'Alto risco', 1: 'Baixo risco'})

# Exibir tabela final com total de famÃ­lias por risco
totais_final = totais[['vulnerabilidade_social', 'total_familias', 'risco_label']]
print(totais_final)
```

```python
import matplotlib.pyplot as plt

# Somar total de famÃ­lias por risco
totais_risco = totais.groupby('risco_label')['total_familias'].sum().reset_index()

# Plotar grÃ¡fico de barras
plt.figure(figsize=(6,4))
plt.bar(totais_risco['risco_label'], totais_risco['total_familias'], color=['red', 'green'])
plt.title('Total de FamÃ­lias por Risco')
plt.xlabel('Risco')
plt.ylabel('NÃºmero de FamÃ­lias')
plt.xticks(rotation=0)
plt.show()
```
